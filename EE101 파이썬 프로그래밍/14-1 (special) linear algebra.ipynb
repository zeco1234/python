# numpy 모듈과 matplotlib 모듈
- numpy.linalg: Linear algebra 
- matplotlib.pyplot: Data visualization 

### 1) scatter

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.random.rand(1000)*6-3 # (-3,3) 사이의 균일 분포
b = np.random.rand(1000)*6-3 
c = np.random.randn(1000) # 정규분포 N(0,1) 
d = np.random.randn(1000) 
plt.scatter(a, b, label='uniform distr')
plt.scatter(c, d, label='Gaussian distr')
plt.legend()
plt.show()
```

### 2) stem

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(10)
x = np.random.rand(10)
y = np.random.rand(10)
plt.stem(n,x, label='x[n]')
plt.stem(y, markerfmt='x', label='y[n]')
plt.legend()
plt.show()
```

### 3) plot

```python
import numpy as np
import matplotlib.pyplot as plt
n = np.linspace(0,1,100)
x = np.cos(4*np.pi*n)
y = np.sin(3*np.pi*n)
plt.plot(n,x, label='$cos[4\pi n]$')
plt.plot(n,y, label='$sin[3\pi n]$')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
y = np.exp(4j*np.pi*n)
plt.plot(np.real(y),'--',color='aqua')
plt.plot(np.imag(y))
plt.show()
```

#### 4) 행렬식의 계산

행렬식 Ax = f 를 계산해보자. 여기서 A는 NXN 행렬이고, f는 N차 벡터이다.

```python
import numpy as np

A = np.array([[3,-1,0,-1,0,0],\
    [-1,2,0,0,-1,0],\
    [0,0,1,0,-2,2],\
    [-3,0,0,1,0,0],\
    [2,-1,0,0,2,-1],\
    [0,0,0,0,-3,5]])
f = np.array([0, 0.006, 0, 0.012, 0, 0])

x = np.linalg.solve(A,f)

y = np.dot(np.linalg.inv(A),f)

z = (np.linalg.inv(A)).dot(f)

print(x)
print(y)
print(z)
```
